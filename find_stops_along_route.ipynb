# Get Stops Along Route

```python
from dotenv import load_dotenv
import pandas as pd
from place_research.places import search_places_nearby
from place_research.routes import find_points_along_route, get_distances_from_origin
from place_research.utils import miles_to_meters, meters_to_miles, seconds_to_hours
from place_research.client import get_client

load_dotenv()
```

```python
subject_address = "1050 Old Mission Rd, New Smyrna Beach, FL 32168"
previous_address = "1050 Old Mission Rd, New Smyrna Beach, FL 32168"
next_address = "1 Hulman Dr, Long Pond, PA 18334"
search_radius_miles = 30
stops_along_route = 3
```

```python
search_radius = miles_to_meters(search_radius_miles)
```

```python
gmaps = get_client()
```

## Find Stops Along a Route

```python
points = find_points_along_route(gmaps, previous_address, next_address, stops_along_route)
    
```

```python
def show_stop_distances(origin: str, points: list):
    distance_points = get_distances_from_origin(gmaps, origin, points)
    
    for point in distance_points:
        distance = meters_to_miles(point['distance'])
        duration = seconds_to_hours(point['duration'])
        print(f"{point['location']} is {distance} miles away and {duration} hours away.")
    
show_stop_distances(previous_address, points)
```
